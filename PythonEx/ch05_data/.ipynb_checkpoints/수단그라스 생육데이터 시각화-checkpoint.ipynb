```python
import matplotlib.pyplot as plt #그래프 패키지 모듈 등록
%matplotlib inline 
#그래프는 show()함수를 통해서 독립창에서 실행되는 것이 원칙
#그래프를 콘솔에서 바로 작도되록 하는 설정
```

```python
# 한글 문제
# matplotlit의 기본 폰트에서 한글 지원되지 않기 때문에
# matplotlib의 폰트 변경 필요
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
import pandas as pd
import numpy as np
```

```python
# 2021 수단그라스 질소처리 data
base_data = pd.read_csv('./data/2021 수단그라스 질소처리 data.csv', encoding='CP949')
base_data
```

```python
base_data.columns
```

```python
plant = base_data[['시기', '처리', '초장(㎝)']]
plant
```

```python
before_plant = plant[plant['시기']=='출수전']
before_plant
```

```python
s1 = before_plant[before_plant['처리']=='N-0']['초장(㎝)']
s2 = before_plant[before_plant['처리']=='N-50']['초장(㎝)']
s3 = before_plant[before_plant['처리']=='N-100']['초장(㎝)']
s4 = before_plant[before_plant['처리']=='N-150']['초장(㎝)']
s5 = before_plant[before_plant['처리']=='N-200']['초장(㎝)']
```

```python
plt.figure(figsize=(6,6))
plt.title('출수전 질소처리 데이터')
plt.boxplot([s1,s2,s3,s4,s5],medianprops=dict(color='k'))
plt.yticks([0,50,100,150,200,250,300,350,400,450])
plt.xticks(range(6),['','N-0','N-50','N-100','N-150','N-200'])
# plt.grid()
plt.show()
```

```python
after_plant = plant[plant['시기']=='출수후']
after_plant

s1 = after_plant[after_plant['처리']=='N-0']['초장(㎝)']
s2 = after_plant[after_plant['처리']=='N-50']['초장(㎝)']
s3 = after_plant[after_plant['처리']=='N-100']['초장(㎝)']
s4 = after_plant[after_plant['처리']=='N-150']['초장(㎝)']
s5 = after_plant[after_plant['처리']=='N-200']['초장(㎝)']

plt.figure(figsize=(6,6))
plt.title('출수후 질소처리 데이터')
plt.boxplot([s1,s2,s3,s4,s5],medianprops=dict(color='k'))
plt.yticks([0,50,100,150,200,250,300,350,400,450])
plt.xticks(range(6),['','N-0','N-50','N-100','N-150','N-200'])
# plt.grid()
plt.show()
```
