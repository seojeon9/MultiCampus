```python
from urllib.request import urlopen # 서버 요청/응답 패키지
import bs4 # 파싱 패키지
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%9A%B0%EC%9D%B4%EB%AA%A9%EC%9E%A5'

# url로 요청 후 응답
html = urlopen(url)

# 파서객체 생성 - bs4 객체로 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```

```python
more = bs_obj.find('a',{'class':'api_more'})
# <a class="api_more" href="https://map.naver.com/v5/entry/place/38431899?c=14140855.609792676%2C4529301.
# 486177189%2C13%2C0%2C0%2C0%2Cdh&amp;placePath=%2F&amp;entry=plt" role="button" target="_blank">
# 우이목장<!-- --> 더보기<i class="spnew ico_arr"></i></a>
more_url = more['href']
```

```python
# url로 요청 후 응답
html = urlopen(more_url)

# 파서객체 생성 - bs4 객체로 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```
