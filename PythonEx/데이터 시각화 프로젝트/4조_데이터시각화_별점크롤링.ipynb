```python
query = df5["업소 전화번호"]
added_query = []
for q in query:
    added_query.append(str(q)+'/')
len(added_query)
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from tqdm import tqdm
```

```python
base_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='
rate = []
for i in tqdm(range(len(added_query))):
    try:
        url = base_url+added_query[i]    
        driver.get(url)
        time.sleep(0.3)
        a = driver.find_element(By.XPATH,'//*[@id="place-main-section-root"]/section/div/div[1]/div[1]/div[2]/span[1]/em')
        rate.append(a.text)
    except:
        rate.append("못찾았음")
    
rate
```
