### 기본은 전화번호로 검색
- 안된다면 검색명으로 전화번호 검색
- 전화번호 전화번호 컬럼에 저장해주고
- 전화번호 + 업소명으로 검색!
- try except 중첩이 안먹히네

```python
url='https://www.naver.com'

driver = webdriver.Chrome('./driver/chromedriver')
driver.get(url)
time.sleep(2)
```

```python
star_list = []
```

```python
for i in tqdm_notebook(res_table.index) :
    if res_table['업소 전화번호'][i] != '-' :
        # 전화번호로 검색
        element = driver.find_element(By.NAME,"query")
        element.clear()
        element.send_keys(res_table['업소 전화번호'][i],' ',res_table['업소명'][i])
        element.send_keys("\n")
        time.sleep(2)
        
        try :
            star_xpath = '//*[@id="place-main-section-root"]/section/div/div[1]/div[1]/div[2]/span[1]/em'
            star = driver.find_element(By.XPATH, star_xpath )
            star_list.append(star.text)
        #   print(star.text)
        except :
            star_list.append('-')
    else :
        # 검색명으로 검색
        element = driver.find_element(By.NAME,"query")
        element.clear()
        element.send_keys(res_table['검색명'][i])
        element.send_keys("\n")
        time.sleep(2)
        
        try :
            star_xpath = '//*[@id="place-main-section-root"]/section/div/div[1]/div[1]/div[2]/span[1]/em'
            star = driver.find_element(By.XPATH, star_xpath )
            star_list.append(star.text)
        #   print(star.text)
        except :
            star_list.append('-')
```
