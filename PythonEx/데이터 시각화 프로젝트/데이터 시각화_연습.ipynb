```python
target_table = res_table[res_table['행정동']==input_dong ]
target_table.index = [i for i in range(len(target_table))]
target_table
```

```python
map_1 = folium.Map(location=[target_table['위도'][0],target_table['경도'][0]],
                   zoom_start=13)

for i in target_table.index :
    name = target_table['업소명'][i]
    star = target_table['별점'][i]
    menu = target_table['메뉴'][i]
    price = target_table['가격'][i]
    url = target_table['식당아이디'][i]

    iframe = "이름 : " + name+ "<br> 별점 : "+ star + "<br> 대표메뉴 :" + menu + " [" + price + "]" + "<br> <a href='https://pcmap.place.naver.com/restaurant/" + url + "/home'>"
    popup = folium.Popup(iframe, min_width=150, max_width=200)
    # 나중에 링크도 추가해도 될 듯
    folium.Marker([target_table['위도'][i],target_table['경도'][i]], 
                  popup=popup, tooltip=name,
                  icon=folium.Icon(color='green')).add_to(map_1)
    
map_1
```
