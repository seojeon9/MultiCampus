```python
df = pd.read_csv("./datas/서대문구 표본(메뉴가격 추가).csv",encoding='cp949',index_col=0)
df.head(2)
df.columns
```

```python
df1 = df[df['상품명'].str.contains('삼겹살') ==True]
df1['메뉴라벨링'] = '삼겹살'
df1['외식비'] = '18056'

df2 = df[df['상품명'].str.contains('냉면') ==True]
df2['메뉴라벨링'] = '냉면'
df2['외식비'] = '10423'

df3 = df[df['상품명'].str.contains('비빔밥') ==True]
df3['메뉴라벨링'] = '비빔밥'
df3['외식비'] = '9654'

df4 = df[df['상품명'].str.contains('김치찌개') ==True]
df4['메뉴라벨링'] = '김치찌개백반'
df4['외식비'] = '7423'

df5 = df[df['상품명'].str.contains('자장면') ==True]
df5['메뉴라벨링'] = '자장면'
df5['외식비'] = '6300'

df6 = df[df['상품명'].str.contains('짜장면') ==True]
df6['메뉴라벨링'] = '자장면'
df6['외식비'] = '6300'

df7 = df[df['상품명'].str.contains('삼계탕') ==True]
df7['메뉴라벨링'] = '삼계탕'
df7['외식비'] = '15385'


df8 = df[df['상품명'].str.contains('칼국수') ==True]
df8['메뉴라벨링'] = '칼국수'
df8['외식비'] = '8385'


df9 = df[df['상품명'].str.contains('김밥') ==True]
df9['메뉴라벨링'] = '김밥'
df9['외식비'] = '2969'
```

```python
df_list = [df1,df2,df3,df4,df5,df6,df7,df8,df9]
```

```python
df_all = pd.concat(df_list, ignore_index=True)
df_all
```

```python
df_all = df_all.sort_values('업소명')
```

```python
df_all.to_csv('./서울시 착한가격음식점(라벨링).csv',encoding='cp949')
```
