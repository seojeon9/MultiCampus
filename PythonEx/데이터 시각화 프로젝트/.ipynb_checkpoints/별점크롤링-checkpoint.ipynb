```python
star_list = []
```

```python
driver = webdriver.Chrome('./driver/chromedriver')
driver.get(url)
time.sleep(2)
```

```python
star_list
```

```python
for res in tqdm_notebook(res_table['업소 전화번호']) :
    element = driver.find_element(By.NAME,"query")
    element.clear()
    element.send_keys(res)
    element.send_keys("\n")
    time.sleep(2)
    
    try :
        star_xpath = '//*[@id="place-main-section-root"]/section/div/div[1]/div[1]/div[2]/span[1]/em'
        star = driver.find_element(By.XPATH, star_xpath )
        star_list.append(star.text)
#         print(star.text)
    except :
        star_list.append('-')
```

```python
print(len(star_list))
star_list.count('-')
```
