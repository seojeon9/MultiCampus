### 카이제곱 검정 개요

- 목적 : 두 범주형 변수가 서로 독립적인지 검정  

    - 귀무가설 : 두 변수가 서로 독립이다
    - 대립가설 : 두 변수가 서로 종속된다  

- 교차테이블로 시각화

#### 교차 테이블(분할표)  

- 두 변수가 취할 수 있는 값의 조합의 출현 빈도를 나타냄  



- 기대값과 실제값의 차이가 클수록 통계량이 커지며 통계량이 커질수록 영가설이 기각될 가능성이 높아짐
