### 일원분산분석 개요

- 목적 : 셋 이상의 그룹 간 차이가 존재하는지를 확인하기 위한 가설 검정 방법
- 영가설 : 세그룹의 표본평균은 같다
- 대립가설 : 최소한 한 개 그룹에는 차이를 보인다 <br><br>



### 선행조건
- 독립성 : 모든 그룹은 서로 독립적이어야 함
- 정규성 : 모든 그룹의 데이터는 정규분포를 따라야 함
    - 그렇지 않으면 비모수적인 방법을 수행해야 함
- 등분산성 : 모든 그룹에 데이터에 대한 분산이 같아야 함
    - 그렇지 않으면 비모수적인 방법을 수행해야 함
