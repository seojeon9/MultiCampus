```python
import pandas as pd
```

```python
dog_kind = pd.read_csv('./품종별 성격.csv', encoding='CP949', index_col=0)
```

```python
dog_kind[dog_kind['품종']!= '미니어처 슈나우저']
```

```python
attraction_high = dog_kind[dog_kind['친화성']=='높음']
attraction_low = dog_kind[dog_kind['친화성']=='낮음']
```

```python
attraction_high_character = []
for words in attraction_high['성격단어']:
    for word in words.split(' '):
        attraction_high_character.append(word)
```

```python
attraction_high_character = set(attraction_high_character)
```

```python
attraction_low_character = []
for words in attraction_low['성격단어']:
    for word in words.split(' '):
        attraction_low_character.append(word)
```

```python
attraction_low_character = set(attraction_low_character)
```

```python
print(attraction_high_character)
print(attraction_low_character)
```

```python
alopecia_high = dog_kind[dog_kind['털빠짐']=='많음']
alopecia_low = dog_kind[dog_kind['털빠짐']=='적음']

alopecia_high_character = []
for words in alopecia_high['성격단어']:
    for word in words.split(' '):
        alopecia_high_character.append(word)
        
alopecia_high_character = set(alopecia_high_character)

alopecia_low_character = []
for words in alopecia_low['성격단어']:
    for word in words.split(' '):
        alopecia_low_character.append(word)
        
alopecia_low_character = set(alopecia_low_character)
```

```python
housekeeping_high = dog_kind[dog_kind['집지키기']=='뛰어남']
housekeeping_low = dog_kind[dog_kind['집지키기']=='부적절']

housekeeping_high_character = []
for words in housekeeping_high['성격단어']:
    for word in words.split(' '):
        housekeeping_high_character.append(word)
        
housekeeping_high_character = set(housekeeping_high_character)

housekeeping_low_character = []
for words in housekeeping_low['성격단어']:
    for word in words.split(' '):
        housekeeping_low_character.append(word)
        
housekeeping_low_character = set(housekeeping_low_character)
```

```python
character_word_df = pd.DataFrame(columns=['피처','정도','성격'])
character_word_df
```

```python
character_word_df.loc[len(character_word_df)] = ['친화성','높음',attraction_high_character]
character_word_df
```

```python
character_word_df.loc[len(character_word_df)] = ['친화성','낮음',attraction_low_character]
character_word_df.loc[len(character_word_df)] = ['털빠짐','많음',alopecia_high_character]
character_word_df.loc[len(character_word_df)] = ['털빠짐','적음',alopecia_low_character]
character_word_df.loc[len(character_word_df)] = ['집지키기','뛰어남',housekeeping_high_character]
character_word_df.loc[len(character_word_df)] = ['집지키기','부적절',housekeeping_low_character]
```

```python
character_word_df.to_csv('./성격단어_샘플.csv',encoding='CP949')
```
