```python
import pandas as pd
from konlpy.tag import Kkma
from konlpy.utils import pprint
```

```python
animal = pd.read_csv('./animal(위경도추가).csv',index_col=0)
```

```python
animal_df = animal[['유기번호','품종','나이','체중','특징']]
```

```python
animal_df['특징'][871]
```

```python
kkma = Kkma()
pos_animal = kkma.pos(animal_df['특징'][13])
```

```python
pos_animal
```

```python
word_list=[]
```

```python
for pos in pos_animal:
    if pos[1] in ['NR','NNM','XR','NNG']:
        word_list.append(pos[0])
```

```python
from tqdm import tqdm_notebook
```

```python
# NNG : 일반 명사 / NNP : 고유 명사 / NR : 수사 / VV : 동사 / VA : 형용사 / MAG : 일반 부사
```

```python
dog_word_list = []
all_words = []

for i in tqdm_notebook(range(len(animal_df))):
    if animal_df['특징'][i].replace(' ','') == '':
        dog_word_list.append('')
    else :
        pos_character = kkma.pos(animal_df['특징'][i])

        word_list = []

        for pos in pos_character:
            if pos[1] in ['NR','NNM','XR','NNG','NNP', 'VV', 'VA', 'MAG']:
                word_list.append(pos[0])
                all_words.append(pos[0])
        word = " ".join(word_list)

        dog_word_list.append(word)
```

```python
dog_word_list
```

```python
animal_df['특징단어'] = dog_word_list
```

```python
animal_df[['특징단어']][:20]
```

```python
all_words
```

```python
all_words_set = set(all_words)
```

```python
len(all_words_set)
```

```python
with open('유기동물 특징 단어.txt','w',encoding='UTF-8') as f:
    for name in all_words_set:
        f.write(name+'\n')
```

```python
animal_df.to_csv('./유기동물별 특징 단어 데이터.csv', encoding='cp949')
```
