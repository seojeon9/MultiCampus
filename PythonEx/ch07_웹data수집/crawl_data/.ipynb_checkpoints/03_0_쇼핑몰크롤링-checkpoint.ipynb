### 쇼핑몰 크롤링
- 상품의 시세 파악을 위해서는 쇼핑몰 데이터를 추출하여 파악하는 것이 일반적임
- 특정 사이트를 선정하여 매일 상품의 가격 변화를 확인
- 사이트 선정 시 구성이 변하지 않고, 접속량이 일정 수준 이상인 사이트를 선정
- http://jolse.com/category/toners-mists/1019/


```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. 소스추출
url = 'http://jolse.com/category/toners-mists/1019/'
htmls = urlopen(url)
htmls = htmls.read()
```

```python
htmls
```

- 요청시 SSL 에러가 발생하면 보안연결을 시도해야 함
- import ssl 패키지를 사용
- ssl 연결 context()를 생성해서 context를 인수로 전달해야 함

```python
# [SSL: CERTIFICATE_VERIFY_FAILED] 에러발생 시

import ssl
context = ssl._create_unverified_context()
url = 'https://jolse.com/category/toners-mists/1019/'
html = urlopen(url, context=context)
htmls = html.read()
```

```python
# 2. 파싱객체 생성
bs_obj = BeautifulSoup(htmls, 'html.parser')
```

**화장품 쇼핑몰**
- 제품명/가격/세일가격
- 페이징 기능을 사용하는 쇼핑몰이므로 페이지를 다르게 접근해서 모든 제품을 크롤링

```python
# 3. 파싱(정보추출)
# 한 페이지에 제품 목록이 2그룹으로 나누어져 있고,
# 각 그룹을 구성하는 태그는 동일
# ul {class:prdList grid5}
uls = bs_obj.findAll('ul',{'class':'prdList grid5'})
len(uls) # 상품진열 목록 2개 추출(추천제품, 일반상품)
```

```python
# 첫번째 진역 목록 - li는 제품정보 내부에 다시 태그가 나타나기 때문에 사용 불가
boxes = uls[0].findAll('li')
len(boxes)
```

```python
boxes = uls[1].findAll('li')
len(boxes)
```

```python
# 상품 정보만 담고 있는 div 태그를 사용(상품 이미지 제외)
# div {class:description}
boxes = uls[0].findAll('div',{'class':'description'})
len(boxes)
```

```python
boxes = boxes + uls[1].findAll('div',{'class':'description'})
len(boxes)
```

### 다른방법
- 모든 상품은 div class description으로 구성되어 있고 해당 태그는 다른 곳에서는 사용되지 않음
- ul에서 내부를 찾지 않고 바로 div를 찾는다


```python
boxes = bs_obj.findAll('div',{'class':'description'})
len(boxes)
```

```python
# 전체 상품명 출력 
for box in boxes :
    strong_tag = box.find('strong',{'class':'name'})
    print(strong_tag.text.split(':')[1])
```

```python
# 정체 상품 리스트 저장
product_list = []
for box in boxes :
    strong_tag = box.find('strong',{'class':'name'})
    product_list.append(strong_tag.text.split(':')[1])
len(product_list)
```

```python
# 가격 정보
boxes[0].find('ul').findAll('span')
# boxes[0].find('ul').findAll('span')[1].text
# 세일가격 : 세일 가격은 없을 수도 있음, -1로 상대적 인덱싱 진행
boxes[0].find('ul').findAll('span')[-1].text # 마지막 span 태그의 텍스트가 추출
```

```python
# 전체 가격 정보 출력 :
for box in boxes :
    # 상품명 출력
    strong_tag = box.find('strong',{'class':'name'})
    print('상품명 : ',strong_tag.text.split(':')[1])
    # 가격
    price = box.find('ul').findAll('span')[1].text
    print('가격 : ',price)
    sale_price = box.find('ul').findAll('span')[-1].text
    print('세일가격 : ',sale_price)
```

### 수집정보 df로 저장

```python
# 빈 리스트 생성
prd_list=[]
price_list=[]
sale_price_list=[]
```

```python
# 전체 상품 데이터 추출 후 저장
boxes = bs_obj.findAll('div',{'class':'description'})
```

```python
for box in boxes :
    # 상품명 출력
    strong_tag = box.find('strong',{'class':'name'})
    prd_list.append(strong_tag.text.split(':')[1])
    # 가격
    price = box.find('ul').findAll('span')[1].text
    price_list.append(price)
    sale_price = box.find('ul').findAll('span')[-1].text
    sale_price_list.append(sale_price)
```

```python
# df생성
prd_dict = {'품목':prd_list,
           '가격':price_list,
           '세일가격':sale_price_list}
prd_df = pd.DataFrame(prd_dict, index=range(1,len(prd_list)+1))
```
